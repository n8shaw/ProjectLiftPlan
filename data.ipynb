```python
import pandas as pd

df = pd.read_csv('C:\\Users\\natha\\OneDrive\\Desktop\\ProjectAlpha\\megaGymDataset.csv')
df.head()
```

```python
def update_gym_column(df):
    for index, row in df.iterrows():
        if pd.isna(row['Equipment']) or row['Equipment'] == 'Body Only':
            df.at[index, 'Gym'] = False
        else:
            df.at[index, 'Gym'] = True

    return df
```

```python
df = update_gym_column(df)

df.head()
```

```python
df.to_pickle('processed_data.pkl')
```
